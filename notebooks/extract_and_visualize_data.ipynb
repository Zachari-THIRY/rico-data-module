# Extracting data from the archive

```python
import sys
if not '../' in sys.path:
    sys.path.append('../')
```

```python
import random
from rico_data_module.rico_raw import RICORawDataset
import pandas as pd
import matplotlib.pyplot as plt
```

## Exporting the desired feature

```python
df = pd.read_hdf('../datasets/RICO1_Dataset.hdf', key='all')
dataset = RICORawDataset(df, ori_seq_len=240, tgt_seq_len=240, stride=1, channels=['B.RTD6'], standardize=False, sampling_rate=10)
```

```python
dataset.export(dir='../datasets', filename='rico1', split=1)
```

See the exported files to see the structure

## Visualizing some samples

```python
idx = [random.randint(0,len(dataset)-1) for i in range(8)]
fig, ax = plt.subplots(2,4, figsize=(15,7))
ax = ax.flatten()

for i in range(8):
    ax[i].plot(dataset[idx[i]])
    ax[i].set_title('Sample N° %d' % idx[i])

plt.show()
```
